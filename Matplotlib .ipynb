# Importing Library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# 1 Line Plot

```python
x=[2,3,4,51,20]
y=[20,30,40,50,40]
plt.plot(x,y)
```

```python
x=[3,5,7,9,11]
y=[4,6,8,10,15]
plt.plot(x,y,'*',color='pink')
plt.title("Line Graph")
plt.xlabel("no of studnt ")
plt.ylabel("marks of studrmt out of 100")
plt.show
```

```python
x = [2 ,5, 6, 7, 8]
y = [3 ,7, 4, 12 ,5]

plt.plot(x, y)

# Formatting the line type  
plt.plot(x,y,'^',color="green") 

plt.title("Line Plot")
plt.ylabel('y-axis')
plt.xlabel('x-axis')
plt.show
```

```python
#sine graph 
x=np.arange(0,3*np.pi,0.1)
y=np.sin(x)

#plotting the line graph for the sine 
plt.plot(x,y,"*",color="yellow")
plt.title("SINE GRAPH")
plt.xlabel("degree ")
plt.ylabel("value")
plt.show
```

```python
#cose graph
a=np.arange(0,3*np.pi,0.2)
b=np.cos(a)
plt.plot(a,b,"*",color="red")
plt.title("cose GRAPH")
plt.xlabel("degree ")
plt.ylabel("value")
plt.show
```

# Scatter Plot

```python
x = [2 ,5, 6, 7, 8]
y = [3 ,7, 4, 12 ,5]

# Formatting the line type  
plt.scatter(x,y,color="green") 

plt.title("Line Plot")
plt.ylabel('y-axis')
plt.xlabel('x-axis')
plt.show
```

```python
x = [2 ,5, 6, 7, 8]
y = [3 ,7, 4, 12 ,5]

plt.scatter(x, y)
plt.title("Scatter Plot")
plt.ylabel('y-axis')
plt.xlabel('x-axis')
plt.show
```

```python
a=np.arange(0,3*np.pi,0.2)
b=np.cos(a)
plt.scatter(a,b,color="red")
plt.title("cose GRAPH")
plt.xlabel("degree ")
plt.ylabel("value")
plt.show
```

```python
x = [2 ,5, 6, 7, 8]
y = [3 ,7, 4, 12 ,5]

x2 = [4 ,5, 8, 9, 8]
y2 = [3 ,5, 7, 12 ,6]

plt.plot(x, y, color = 'g', linewidth = 6, label = 'first')
plt.plot(x2, y2, color = 'r', linewidth = 4, label = 'second')

plt.title("Comparison Graph")
plt.ylabel('y-axis')
plt.xlabel('x-axis')

# Adding Notation
plt.annotate(xy=[2, 3], s='First Line') 
plt.annotate(xy=[4,4], s='Second Line') 

# Adding LEGENDS
plt.legend(['First Line', 'Second Line'], loc=4)

plt.show
```

```python
x = [2 ,5, 6, 7, 8]
y = [3 ,7, 4, 12 ,5]

x2 = [4 ,5, 8, 9, 8]
y2 = [3 ,5, 7, 12 ,6]

plt.bar(x, y, color = 'r', linewidth = 5, label = 'first')
plt.bar(x2, y2, color = 'y', linewidth = 5, label = 'second')

plt.title("Bar Graph")
plt.ylabel('y-axis')
plt.xlabel('x-axis')

# Adding LEGENDS
plt.legend(['First Line', 'Second Line'], loc=5)
plt.show
```

# Horizontal Bar Graph

```python
x = [2 ,5, 6, 7, 8]
y = [3 ,7, 4, 12 ,5]

x2 = [4 ,5, 8, 9, 8]
y2 = [3 ,5, 7, 12 ,6]

plt.barh(x, y, color = 'r', linewidth = 5, label = 'first')
plt.barh(x2, y2, color = 'y', linewidth = 5, label = 'second')

plt.title("Bar Graph")
plt.ylabel('y-axis')
plt.xlabel('x-axis')

# Adding LEGENDS
plt.legend(['First Line', 'Second Line'], loc=10)
plt.show
```

# Bubble Graph

```python
x = [2 ,5, 6, 7, 8]
y = [3 ,7, 4, 12 ,5]

plt.scatter(x, y,color="pink")
plt.title("Scater Plot")
plt.ylabel('y-axis')
plt.xlabel('x-axis')
plt.show
```

```python
labels = ['a', 'b', 'c', 'd', 'e']

# Exploding 2nd slice
explode = (0, 0.1, 0.1, 0, 0)

data = [3 ,7, 4, 12 ,5]
plt.pie(data, explode=explode, labels = labels, shadow = True)
plt.show()
```
